<p>  Import necessary packages </p>

```python
import pandas as pd
import matplotlib.pyplot as plot
from collections import Counter
```

<p> Answer the questions and save as csv file</p>

```python

dataframe = pd.read_csv('imdb_top_1000.csv')
columns = dataframe.columns
imdbRating = dataframe['IMDB_Rating']
genres = dataframe['Genre']
averageRating = imdbRating.mean()
movieGenres = [genre.split(',') for genre in genres]
allGenres = [singleGenre.strip() for genres in movieGenres for singleGenre in genres]
uniqueGenres = Counter(allGenres)
answer = pd.DataFrame({
    'Columns': [columns],
    'Average_Rating': averageRating,
    'Unique_Genres': len(uniqueGenres)
})


answer.to_csv('answer.csv', index=False)

```

<p>Plot histogram of each genres frequency in the top 1000 IMBD Movies</p>

```python
plot.figure(figsize=(40, 10))  
plot.hist(allGenres,bins=len(uniqueGenres), edgecolor='darkblue')
plot.xlabel('Genres')
plot.ylabel('Frequency')
plot.title('Genres frequency in IMDB top 1000 movies')
plot.show()
```

<p>scatter plot of how many movies made to the top 1000 list each year</p>

```python
topThousandPerYear = Counter(dataframe['Released_Year'])
currentYear = 2023
startYear = int(min(topThousandPerYear.keys(), key= lambda year: int(year) if year.isdigit() else currentYear))
years = [ str(year) for year in range(startYear, currentYear) ]

plot.figure(figsize=(60, 20))  
plot.scatter(years, [topThousandPerYear[year] for year in years], )
```
