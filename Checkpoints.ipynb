**CP1**:
Given the list $[A, B, C]$, use the `itertools` module to generate all possible
combinations of 2 elements.

**Expected Output**:
$$
 (A, B) \\
 (A, C) \\
 (B, C)
$$

```python
#your code here
letters = ['A','B','C']
for combination in  combinations(letters, 2):
    print(combination)
```

**CP2**:
Use the `itertools` module to create a sequence that cycles through
['Python', 'is', 'fun'] infinitely. Print the first 10 elements of the sequence.

**Expected Output**:
$$
Python \\
is\\
fun\\
Python\\
is\\
fun\\
Python\\
is\\
fun\\
Python
$$

```python
#your code here
rng = 10
for word in cycle(['Python', 'is', 'fun']):
    if rng <= 0:
        break
    print(word)
    rng -= 1
    
```

**CP4**: You are given the following HTML content below. Write a function that uses BeautifulSoup to parse the HTML and extract the text inside the   '$<p>$'   tag.


```python
html = "<html><body><div class='main'><h1>Welcome to AI</h1><p>AI is transforming the world</p></div></body></html>"
```

```python
#your code here
def extractP(html):
    soup = BeautifulSoup(html, 'html.parser')
    return soup.p.get_text()

extractP(html)
```

**CP5**: Write a function that uses the re module to replace all occurrences of the word "AI" with "Artificial Intelligence" in a given text.

```python
#your code here
text = 'AI is a transformative field of study that leverages advanced algorithms and computational models to enable machines to mimic human intelligence and perform tasks that traditionally required human cognition. With its ability to analyze vast amounts of data, AI has revolutionized industries such as healthcare, finance, and transportation, powering innovations like autonomous vehicles, personalized medicine, and intelligent virtual assistants. From machine learning and natural language processing to computer vision and neural networks, AI encompasses a wide range of techniques that enable systems to learn, reason, and make decisions autonomously, paving the way for a future where AI plays a pivotal role in enhancing efficiency, solving complex problems, and driving unprecedented progress across numerous domains.'
original_word = 'AI'
new_word = 'Artificial Intelligence'
def replaceWord(original_word,new_word,  text):
    return re.sub(original_word, new_word, text)

replaceWord(original_word=original_word, new_word=new_word, text=text)
```
